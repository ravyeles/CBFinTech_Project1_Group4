```python
#Simulate Weights
numOfAssets=5
trials = 500
overall_weights=[]
total = 0
for i in range(trials):
    weights=[]
    for i in range(numOfAssets):
        weights.append(random.random())
    total = sum(weights)
    for i in range(len(weights)):
        weights[i]=round(weights[i]/total,2)
    overall_weights.append(weights)
```

```python
#Create a dataframe to store the results of the MC Simulations for each set of weights
mc_returns = pd.DataFrame(columns=['weights','mean','median','min','max','95% CI Lower','95% CI Upper','std_dev'])
#create an index for the dataframe
mc_returns.insert(0,'index',range(0,trials))
#Specify the data types for the columns
mc_returns.astype({'weights': 'object', 'mean': 'float64', 'median': 'float64', 'min': 'float64', 'max': 'float64', '95% CI Lower': 'float64', '95% CI Upper': 'float64', 'std_dev': 'float64'})
#initialize a counter to keep track of the row in the dataframe
counter = 0

#Simulate Weights for use in the MC Simulation
import random 
numOfAssets=2
combinations = 500
overall_weights=[]
total = 0
for i in range(combinations):
    weights=[]
    for i in range(numOfAssets):
        weights.append(random.random())
    total = sum(weights)
    for i in range(len(weights)):
        weights[i]=round(weights[i]/total,2)
    overall_weights.append(weights)

#iterate through the weights and run the MC Simulation for each set of weights
for weight in overall_weights:
    #Store the weight being used in the simulation
    mc_returns.at[counter, 'weights'] = weight
    #Run the MC Simulation and store the results in the dataframe
    MC_fiveyear = MCSimulation(
        portfolio_data = df_ticker,
        weights = weight,
        num_simulation = 500,
        num_trading_days = 252*5
    )
    # Run the Monte Carlo simulation to forecast five years cumulative returns
    MC_fiveyear.calc_cumulative_return()
    # Summarize the results of the Monte Carlo simulation
    tbl = MC_fiveyear.summarize_cumulative_return()
    #Store the results in the dataframe
    mc_returns.at[counter, 'mean'] = float(tbl['mean'])
    mc_returns.at[counter, 'median'] = float(tbl['50%'])
    mc_returns.at[counter, 'min'] = float(tbl['min'])
    mc_returns.at[counter, 'max'] = float(tbl['max'])
    mc_returns.at[counter, '95% CI Lower'] = float(tbl['95% CI Lower'])
    mc_returns.at[counter, '95% CI Upper'] = float(tbl['95% CI Upper'])
    mc_returns.at[counter, 'std_dev'] = float(tbl['std'])
    #increment the counter
    counter += 1

#Determine which record has the highest 95% CI Upper value and store that in max_return dataframe
max_returns = mc_returns[mc_returns['95% CI Upper']==mc_returns['95% CI Upper'].max()]
#Determine which record has the lowest standard deviation and store that in min_volatility dataframe
min_volatility = mc_returns[mc_returns['std_dev']==mc_returns['std_dev'].min()]
# Determine which record has the highest average return and store that in the balanced dataframe
balanced = mc_returns[mc_returns['mean']==mc_returns['mean'].max()]
```
